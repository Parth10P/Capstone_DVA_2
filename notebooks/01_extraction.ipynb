## Data Extraction

Dataset: Online Retail (UCI)
Link: https://archive.ics.uci.edu/dataset/352/online+retail

Downloaded the xlsx file and placed it in data/raw/. Not touching that file at all.

```python
import pandas as pd

df = pd.read_excel('../data/raw/Online Retail.xlsx')
print(df.shape)
df.head()
```

```python
# quick look at columns and types
print(df.dtypes)
print()
print(df.isnull().sum())
```

```python
# how many unique invoices and customers
print('Invoices:', df['InvoiceNo'].nunique())
print('Customers:', df['CustomerID'].nunique())
print('Countries:', df['Country'].nunique())
```

```python
# date range
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print('From:', df['InvoiceDate'].min())
print('To:  ', df['InvoiceDate'].max())
```
